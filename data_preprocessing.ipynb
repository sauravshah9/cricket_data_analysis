```python
import pandas as pd
import json
```

```python
df = pd.read_csv("match_data.csv")
df
```

```python
new_df  = df.drop('id',axis=1)
```

```python
new_df.shape
```

```python
new_df.rename(columns = {'scorecard':'match_id'}, inplace = True)
```

```python
new_df.head()
```

```python
with open("t20_json_files/t20_wc_batting_summary.json") as f:
    data = json.load(f)
    all_rec =[]
    for rec in data:
        all_rec.extend(rec['battingSummary'])
# all_rec

df_batting = pd.DataFrame(all_rec)
df_batting
```

```python
match_id_dict={}

for index,row in new_df.iterrows():
    key1 = row['team1'] + " Vs " +row['team2']
    key2 = row['team2'] + " Vs " +row['team1']   
    
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']    

match_id_dict
```

```python
df_batting["out/not_out"] = df_batting["dismissal"].apply(lambda x: "out" if len(x)>0 else "not_out")
```

```python
df_batting.head(11)
```

```python
df_batting.drop("dismissal",axis= 1,inplace=True)
```

```python
df_batting.head(11)
```

```python
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x: x.replace("â€",""))
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x: x.replace("\xa0",""))
```

```python
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting["match"].map(match_id_dict)
```

```python
df_batting.head(50)
```

```python
df_batting.to_csv("t20_csv_files/fact_bating_summary.csv",index = False)
```

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```
